```python
df['PhoneNumber'] = df['PhoneNumber'].map(lambda x: ''.join([i for i in x if i.isdigit()]))
df['PhoneNumber']
```

```python
df_10digit = df.loc[df['PhoneNumber'].str.len()==10]
```

```python
a=[]
for i in df_10digit['PhoneNumber']:
    reformat_phone = "(%c%c%c)%c%c%c-%c%c%c%c" % tuple(map(ord,list(str(i)[:10])))
    a.append(reformat_phone)
print(a)
```

```python
df_reformat = pd.DataFrame()
df_reformat['PhoneNumber']=a
```

```python
df_reformat
```

```python
df_reformat.to_csv('PhoneNumber.csv',index=False)
```
